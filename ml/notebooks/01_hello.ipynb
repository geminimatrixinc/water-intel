```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a tiny dummy dataframe
data = {
    'x': [1, 2, 3, 4, 5],
    'y': [2, 4, 6, 8, 10]
}
df = pd.DataFrame(data)
df
```

```python
# Plot a line chart
plt.figure(figsize=(8, 5))
plt.plot(df['x'], df['y'], marker='o', linewidth=2, label='Data')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Simple Line Chart')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
